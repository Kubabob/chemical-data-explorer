```python
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors
import pandas as pd
```

```python
Mocap = Chem.MolFromSmiles('CCCSP(=O)(OCC)SCCC')
Draw.MolToImage(Mocap)
```

```python
DDT = Chem.MolFromSmiles('C(Cl)(Cl)(Cl)C(c1ccc(Cl)cc1)(c2ccc(Cl)cc2)')
Draw.MolToImage(DDT)
```

```python
Frumin = Chem.MolFromSmiles('CCSCCSP(OCC)(OCC)=S')
Draw.MolToImage(Frumin)
```

```python
Rampart = Chem.MolFromSmiles('CCSCSP(OCC)(OCC)=S')
Draw.MolToImage(Rampart)
```

```python
Thiacloprid = Chem.MolFromSmiles('N#CN=C1SCCN1CC2=C-N=C(Cl)C=C2')
Draw.MolToImage(Thiacloprid)
```

```python
smiles_codes = pd.read_excel('Smiles_codes.xlsx')
smiles_codes['Smiles']
```

```python
mols = []
descriptors = []
for smile in smiles_codes['Smiles']:
    mol = Chem.MolFromSmiles(smile)
    mols.append(mol)
    descriptors.append(Descriptors.CalcMolDescriptors(mol))
    
descriptors = pd.DataFrame(descriptors)
descriptors.to_excel('Descriptors.xlsx')
print(descriptors)
Draw.MolToImage(mols[4])
```
