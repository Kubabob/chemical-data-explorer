```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
pd.set_option('mode.chained_assignment', None)
pd.set_option('future.infer_string', False)
```

```python
descriptors = pd.read_excel('Descriptors.xlsx', index_col=0)
descriptors = descriptors.dropna()
descriptors = descriptors.astype(float)
descriptors
```

```python
for column in descriptors:
    if descriptors[column].var() == 0:
        descriptors = descriptors.drop(column, axis=1)


descriptors
```

```python
for column_idx, column in enumerate(descriptors):
    mean = descriptors[column].mean()
    std = descriptors[column].std()
    for idx in descriptors.index:
        descriptors.loc[idx, column] = float((descriptors.loc[idx, column] - mean)/std)

descriptors = descriptors.dropna()
descriptors
```

```python
descriptors_odl_eukl = np.zeros([descriptors.shape[0] for _ in range(2)])

for w in range(descriptors.shape[0]):
    for k in range(descriptors.shape[0]):
        descriptors_odl_eukl[w,k] = np.sqrt(sum((descriptors.T[descriptors.T.columns[w]] - descriptors.T[descriptors.T.columns[k]])**2))


descriptors_odl_eukl = pd.DataFrame(descriptors_odl_eukl, columns=descriptors.index, index=descriptors.index)
descriptors_odl_eukl.to_excel('Odleglosc_euklidesowa.xlsx')
descriptors_odl_eukl
```

```python
descriptors_odl_manhattan = np.zeros([descriptors.shape[0] for _ in range(2)])

for w in range(descriptors.shape[0]):
    for k in range(descriptors.shape[0]):
        descriptors_odl_manhattan[w,k] = sum(abs((descriptors.T[descriptors.T.columns[w]] - descriptors.T[descriptors.T.columns[k]])))


descriptors_odl_manhattan = pd.DataFrame(descriptors_odl_manhattan, columns=descriptors.index, index=descriptors.index)
descriptors_odl_manhattan.to_excel('Odleglosc_manhattan.xlsx')
descriptors_odl_manhattan
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
linkage_data = linkage(descriptors, 'single', 'euclidean')
print(linkage_data[-1][2])
dendrogram(linkage_data, labels=descriptors.index, leaf_rotation=90)
plt.plot([0,1000], [linkage_data[-1][2]*2/3 for _ in range(2)], linestyle='dashed')
plt.show()
```

```python
def HCA(distance_matrix: pd.DataFrame, linkage):
    alphabet = iter(list(map(chr, range(97, 123))))
    linkedValCoordsList = []
    for _ in range(distance_matrix.shape[0]-1):
        linkedValue = distance_matrix[distance_matrix > 0].min(axis=0).min()
        linkedValCoords = list(list(zip(*np.where(distance_matrix==linkedValue)))[0])
        linkedValCoords[0], linkedValCoords[1] = distance_matrix.index[linkedValCoords[0]], distance_matrix.columns[linkedValCoords[1]]
        linkedValCoordsList.append([linkedValCoords, linkedValue])

        if linkage == 'single':
            newRowVector, newColumnVector = np.min([distance_matrix.loc[linkedValCoords[0]], distance_matrix.loc[linkedValCoords[1]]], axis=0), np.min([distance_matrix.loc[:,linkedValCoords[0]], distance_matrix.loc[:,linkedValCoords[1]]], axis=0)
        elif linkage == 'complete':
            newRowVector, newColumnVector = np.max([distance_matrix.loc[linkedValCoords[0]], distance_matrix.loc[linkedValCoords[1]]], axis=0), np.max([distance_matrix.loc[:,linkedValCoords[0]], distance_matrix.loc[:,linkedValCoords[1]]], axis=0)        

        nextLetter = next(alphabet)
        distance_matrix = distance_matrix.rename(columns={linkedValCoords[0]:nextLetter}, index={linkedValCoords[0]:nextLetter})


        distance_matrix.loc[nextLetter], distance_matrix.loc[:,nextLetter] = newRowVector, newColumnVector

        distance_matrix.loc[nextLetter, nextLetter] = 0
        
        distance_matrix = distance_matrix.drop(linkedValCoords[1],axis=0)
        distance_matrix = distance_matrix.drop(linkedValCoords[1],axis=1)

    return linkedValCoordsList
```

```python
HCA(descriptors_odl_eukl, 'single')
```
