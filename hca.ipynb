```python
age = np.array([-0.24, -1.38, 0.62, -0.9, 1.09, 0.81])
height = np.array([-0.7, -1.45, 0.35, 1.49, 0.06, 0.25])
weight = np.array([-0.68, -1.35, 0.68, 1.5, -0.08, -0.08])
bmi = np.array([-0.55, -1.31, 1.15, 1.23, -0.14, -0.38])



m_ahwb = np.array([age,height,weight,bmi]).T
m_ahwb
```

```python
m_odl = np.zeros([max(m_ahwb.shape) for _ in range(2)])
m_odl
```

```python
for k in range(m_odl.shape[0]):
    for w in range(m_odl.shape[0]):
        m_odl[w,k] = np.sqrt(sum((m_ahwb[k] - m_ahwb[w])**2))
print(m_odl)
```

```python
import pandas as pd
```

```python
descriptors = pd.read_excel('Descriptors.xlsx')
descriptors.dropna()
```

```python
for column in descriptors:
    if descriptors[column].var() == 0:
        descriptors = descriptors.drop(column, axis=1)

descriptors = descriptors.drop('Unnamed: 0', axis=1)

descriptors
```

```python
for column in descriptors:
    for index, value in enumerate(descriptors[column]):
        descriptors[column][index] = (value - descriptors[column].mean())/descriptors[column].std()

descriptors
```

```python
descriptors_odl = np.zeros([descriptors.shape[0] for _ in range(2)])

for w in range(descriptors.shape[0]):
    for k in range(descriptors.shape[0]):
        descriptors_odl[w,k] = np.sqrt(sum((descriptors[descriptors.columns[w]] - descriptors[descriptors.columns[k]])**2))

print(descriptors_odl)
```
