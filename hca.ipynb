```python
import pandas as pd
pd.set_option('mode.chained_assignment', None)
pd.set_option('future.infer_string', False)
```

```python
descriptors = pd.read_excel('Descriptors.xlsx')
descriptors.dropna()
```

```python
for column in descriptors:
    if descriptors[column].var() == 0:
        descriptors = descriptors.drop(column, axis=1)

descriptors = descriptors.drop('Unnamed: 0', axis=1)

descriptors
#print(descriptors.dtypes)
```

```python
for column in descriptors:
    mean = descriptors[column].mean()
    std = descriptors[column].std()
    for index, value in enumerate(descriptors[column]):
        descriptors[column][index] = float((value - mean)/std)

descriptors
```

```python
descriptors_odl_eukl = np.zeros([descriptors.shape[0] for _ in range(2)])

for w in range(descriptors.shape[0]):
    for k in range(descriptors.shape[0]):
        descriptors_odl_eukl[w,k] = np.sqrt(sum((descriptors.T[descriptors.T.columns[w]] - descriptors.T[descriptors.T.columns[k]])**2))


pd.DataFrame(descriptors_odl_eukl).to_excel('Odleglosc_euklidesowa.xlsx')
pd.DataFrame(descriptors_odl_eukl)
```

```python
descriptors_odl_manhattan = np.zeros([descriptors.shape[0] for _ in range(2)])

for w in range(descriptors.shape[0]):
    for k in range(descriptors.shape[0]):
        descriptors_odl_manhattan[w,k] = sum(abs((descriptors.T[descriptors.T.columns[w]] - descriptors.T[descriptors.T.columns[k]])))

pd.DataFrame(descriptors_odl_manhattan).to_excel('Odleglosc_manhattan.xlsx')
pd.DataFrame(descriptors_odl_manhattan)
```
